# Filling in Missing Values

```python
# Added a name to a column within the CSV file 'Date'
dfEconomy = pd.read_csv('Modified Data/EconomicIndicators.csv', index_col='Date')
```

```python
#Calculating Constant GNI (2015) for missing values
#Constant GNI = Current GNI / Deflator
for i in range(1960, 1986):
    GDP_Deflator = dfEconomy.loc[i, 'GDP (current US$)'] / dfEconomy.loc[i, 'GDP (constant 2015 US$)']
    dfEconomy.loc[i, 'GNI (constant 2015 US$)'] = dfEconomy.loc[i, 'GNI (current US$)'] / GDP_Deflator
```

```python
#Calculating GNI Growth (it was originally calculated using Constant GNI so we'll use that)
#Growth = (GNI 2 / GNI 1) - 1
for i in range(1961, 1987):
    dfEconomy.loc[i, 'GNI growth (annual %)'] = ((dfEconomy.loc[i, 'GNI (constant 2015 US$)'] / dfEconomy.loc[i-1, 'GNI (constant 2015 US$)']) - 1) * 100
```

```python
#Calculating Current Exports and Imports
#Current Exports = Constant Exports * Deflator
for i in range(1960, 1976):
    GDP_Deflator = dfEconomy.loc[i, 'GDP (current US$)'] / dfEconomy.loc[i, 'GDP (constant 2015 US$)']
    dfEconomy.loc[i, 'Exports of goods and services (BoP, current US$)'] = dfEconomy.loc[i, 'Exports of goods and services (constant 2015 US$)'] * GDP_Deflator
    dfEconomy.loc[i, 'Imports of goods and services (BoP, current US$)'] = dfEconomy.loc[i, 'Imports of goods and services (constant 2015 US$)'] * GDP_Deflator
```

```python
#Calculating GDP PPP Growth (using constant GDP PPP)
#Growth = (GDP 2 / GDP 1) - 1
dfEconomy['GDP, PPP Growth (annual %)'] = np.nan
for i in range(1991, 2022):
    dfEconomy.loc[i, 'GDP, PPP Growth (annual %)'] = ((dfEconomy.loc[i, 'GDP, PPP (constant 2017 international $)'] / dfEconomy.loc[i-1, 'GDP, PPP (constant 2017 international $)']) - 1) * 100
```
